```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('dark_background')
```

```python
x = np.linspace(0,4*np.pi,200)
y1_1 = np.sin(x)
y2_1 = np.sin(x + 1)
y3_1 = y1_1 + y2_1
y1_2 = np.sin(x)
y2_2 = np.sin(x + 0.05)
y3_2 = y1_2 + y2_2
y1_3 = np.sin(x)
y2_3 = np.sin(x + np.pi)
y3_3 = y1_3 + y2_3
```

```python
fig = plt.figure(dpi=300)
ax = fig.add_subplot(1, 1, 1)

# Move left y-axis and bottim x-axis to centre, passing through (0,0)
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('center')

# Eliminate upper and right axes
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# Show ticks in the left and lower axes only
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

# set ticks
# ax.xaxis.set_ticks(np.pi*np.arange(0.5,4.0,0.5))
# ax.xaxis.set_ticks(np.pi*np.arange(0.5,4.0,0.5))
# ax.xaxis.set_ticklabels([r'$\pi/2$',r'$\pi$',r'$3 \pi /2$',r'$2 \pi$',r'$5 \pi /2$',r'$3 \pi$',r'$7 \pi /2$'])
ax.yaxis.set_ticklabels([])

# axis labels
ax.set_xlabel('$x$', loc='right')
ax.set_ylabel('$y(x)$', loc='top')
ax.plot(x,y1_1,'--',
        x,y2_1,':',
        x,y3_1)

ax.legend(['$y_{1}(x,t)$','$y_{2}(x,t)$','$y_{1}(x,t) + y_{2}(x,t)$'])
ax.set_title(r'$\phi=\phi_{0}$',fontsize=20)
plt.tight_layout()
fig.savefig('../wave_interference_1.png')
```
